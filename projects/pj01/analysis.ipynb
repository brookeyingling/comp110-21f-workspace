```python
from data_utils import helper_function 
zero: float = helper_function(selected_data, "oh_visits", "difficulty", "0")
one: float = helper_function(selected_data, "oh_visits", "difficulty", "1")
two: float = helper_function(selected_data, "oh_visits", "difficulty", "2")
three: float = helper_function(selected_data, "oh_visits", "difficulty", "3")
four: float = helper_function(selected_data, "oh_visits", "difficulty", "4")
five: float = helper_function(selected_data, "oh_visits", "difficulty", "5")
empty_dictionary: dict[str, float] = {}
empty_dictionary["0"] = zero
empty_dictionary["1"] = one
empty_dictionary["2"] = two
empty_dictionary["3"] = three
empty_dictionary["4"] = four
empty_dictionary["5"] = five
print(empty_dictionary)
```
